```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D


from sklearn import decomposition
from sklearn import datasets
```

```python
ds = pd.read_csv("dataset\S01R02.csv")
ds1 = ds[ds["Annotation [0/1/2]"] == 1]
ds2 = ds[ds["Annotation [0/1/2]"] == 2]
ds.head()
```

```python
X1 = ds1[ds1.columns[1:-1]]
y1 = ds1["Annotation [0/1/2]"]


fig = plt.figure()
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

def scatter_plot(X,y, param_color,s1):
    
    pca = decomposition.PCA(n_components=3)
    pca.fit(X)
    X = pca.transform(X)

    ax.scatter(X[:, 0], X[:, 1], X[:, 2],  c=param_color ,s=s1)

```

```python
scatter_plot(X1, y1, 'blue',0.1)
```
