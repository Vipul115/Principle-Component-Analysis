```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
ds = pd.read_csv("dataset\S01R02.csv")
ds.head()
```

```python
ds.columns
```

```python
ds2 = ds[ds['Annotation [0/1/2]'] == 2]
```

```python
ds1 = ds[ds['Annotation [0/1/2]'] ==1]
```

```python
def plotter(feature):
    plt.figure()
    ax = plt.subplot(111)
    bars1  = plt.hist(ds1[feature], bins=np.linspace(-5000, 5000, 20) , color='yellow', label="Non-Freezes")
    bars2  = plt.hist(ds2[feature], bins=np.linspace(-5000, 5000, 20) ,color='blue', alpha=0.6, label="Freezes")
    plt.xlabel(feature)
    plt.ylabel("Frequency")
    plt.xticks(np.arange(-5000, 5000, 1000), rotation=90)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    # plt.subplots_adjust(bottom=0.2)
    plt.tight_layout()

    plt.legend()
    plt.show()
for col in ds.columns[1:]:
    plotter(col)
```
